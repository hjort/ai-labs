```python
import pandas as pd
import numpy as np
```

```python
# carregar dados
data = pd.read_csv('data/winequality-white.csv', sep=';')

# embaralhar dados
data = data.sample(frac=1)

# mostrar quantidade de linhas e colunas
print(data.shape)

# mostrar exemplos de dados
data.head()
```

```python
# renomear coluna índice
data.index.names = ['wine']
```

```python
# modificar nomes das colunas

cols = {}
for col in data.columns:
    newcol = col.lower().replace(' ', '_')
    cols[col] = newcol
    #print(col, '=>', newcol)

#print(type(cols), cols)

data.rename(columns=cols, inplace=True)
```

```python
data.quality = data.quality.map(lambda x: 'good' if x > 6 else 'bad')
```

```python
data['quality'].describe()
```

```python
# exibir dados
data.head()
```

```python
# dividir os dados de treino e teste

divisao = int(data.shape[0] * 2 / 3)

train = data[:divisao]
test = data[divisao:]

print(train.shape, test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
example = test[['quality']]
#example['quality'] = example.index.map(lambda x: float(x % 7 + 3))
example['quality'] = example.index.map(lambda x: 'good' if (x % 2 == 0) else 'bad')
example.head()
```

```python
example.describe()
```

```python
train.to_csv('wine-train.csv')

test.drop(['quality'], axis=1).to_csv('wine-test.csv')
test[['quality']].to_csv('wine-solution.csv')

example.to_csv('wine-example.csv')
```
