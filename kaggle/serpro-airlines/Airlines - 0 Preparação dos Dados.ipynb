```python
import pandas as pd
import numpy as np
```

```python
# definir parâmetros extras
pd.set_option('precision', 4)
pd.set_option('display.max_columns', None)
```

```python
# carregar dados
data = pd.read_csv('data/allyears2k_headers.csv', sep=',', encoding='ISO-8859-1')

# embaralhar dados
data = data.sample(frac=1)

# mostrar quantidade de linhas e colunas
print(data.shape)

# mostrar exemplos de dados
data.head()
```

```python
# renomear coluna índice
data.index.names = ['FlightRecord']
```

```python
data.info()
```

```python
data.describe()
```

```python
data.describe(include=['O'])
```

```python
# criar coluna única com informação da data
data['FlightDate'] = pd.to_datetime(data[['Year', 'Month', 'DayofMonth']].rename(columns={'DayofMonth': 'Day'}))
```

```python
# remover colunas derivadas da data
data.drop(['Year', 'Month', 'DayofMonth', 'DayOfWeek'], axis=1, inplace=True)
```

```python
# arrumar ordem das colunas
cols = data.columns.values[:-1].tolist()
cols.insert(0, 'FlightDate')
data = data[cols]
```

```python
# dividir os dados de treino e teste
divisao = int(data.shape[0] * 4 / 5)
train = data[:divisao]
test = data[divisao:]
print(train.shape, test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
example = test[['IsDepDelayed']]
example['IsDepDelayed'] = example.index.map(lambda x: 'NO' if (x % 2 == 0) else 'YES')
example.head()
```

```python
example.describe()
```

```python
# gerar arquivos CSV
train.to_csv('airlines-train.csv')
test.drop(['IsDepDelayed'], axis=1).to_csv('airlines-test.csv')
test[['IsDepDelayed']].to_csv('airlines-solution.csv')
example.to_csv('airlines-example.csv')
```
