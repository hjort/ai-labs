```python
import pandas as pd
import numpy as np
```

```python
# definir par√¢metros extras
pd.set_option('precision', 4)
pd.set_option('display.max_columns', 100)
```

```python
# carregar dados
data = pd.read_csv('temperature.csv', sep=',', index_col='datetime')

# mostrar quantidade de linhas e colunas
print(data.shape)

# mostrar exemplos de dados
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# manter apenas uma cidade
data = data[['Detroit']]

# renomear colunas
data.rename(columns={'Detroit': 'temperature'}, inplace=True)
```

```python
# converter temperatura para graus Celsius
data['temperature'] = round(data['temperature'].apply(lambda k: k - 273.15), 4)
```

```python
data.info()
```

```python
data.describe()
```

```python
# exibir dados
data.head()
```

```python
# dividir os dados de treino e teste

divisao = int(data.shape[0] * 4 / 5)

train = data[:divisao]
test = data[divisao:]

print(train.shape, test.shape)
```

```python
train.head()
```

```python
test.head()
```

```python
example = test[['temperature']]

# gerar valores aleat√≥rios de temperatura (m√≠n: -30¬∫C, m√°x: 35¬∫C)
example['temperature'] = np.round(np.random.random(example.shape[0]) * 65 - 30, 4)

example.head()
```

```python
example.describe()
```

```python
train.to_csv('weather-train.csv')

test.drop(['temperature'], axis=1).to_csv('weather-test.csv')
test[['temperature']].to_csv('weather-solution.csv')

example.to_csv('weather-example.csv')
```
