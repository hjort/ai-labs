```python
# importar pacotes necessários
import numpy as np
import pandas as pd
```

```python
# definir parâmetros extras
pd.set_option('precision', 4)
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
input_prefix = 'data/'
#input_prefix = 'https://github.com/hjort/ai-labs/raw/master/jupyter/sales-russia/data/'

bzfile = '.bz2' # ''
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%d.%m.%Y')
df_train = pd.read_csv(\
    input_prefix + 'sales_train.csv' + bzfile,
    #nrows=300, # somente primeiras linhas!
    parse_dates=['date'], date_parser=dateparse,
    dtype={'date_block_num': np.int8, 'shop_id': np.int8, 'item_id': np.int16, \
           'item_price': np.float32, 'item_cnt_day': np.int32},
    )
print('shape:', df_train.shape)
df_train.head()
```

```python
%time df_train['item_price'] = (np.ceil(df_train['item_price'])).astype(np.int32)
```

```python
grouped_data = df_train.groupby(['shop_id', 'item_id'], as_index=False).agg(
    {'item_price': ['count', 'sum', 'min', 'max', 'mean', 'std', 'median'],
     'item_cnt_day': ['count', 'sum', 'min', 'max', 'mean', 'std', 'median']}
)

grouped_data.sort_values(by=('item_price', 'count'), ascending=False).head(15)
```
