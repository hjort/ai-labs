Teste com Alpha = 5%

```python
zconfint(notas1.rating)
```

O Teste Z é mais recomendado quando estamos trabalhando com uma amostra expressiva de dados!

```python
from statsmodels.stats.weightstats import ztest

# retorna t-test e p-value respectivo
ztest(notas1.rating, value=3.4320503405352594)
```

p-value = 8.97e-18, que é menor que 5%, portanto a hipótese nula é descartada!

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value = 3.4320503405352594)
  return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:,0], valores[:,1]) # qtd. amostras vs z-test
plt.plot(valores[:,0], valores[:,2]) # qtd. amostras vs p-value
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors='r') # linha 5%
```

A quantidade de 50 amostras é um tamanho que passa a ser suficiente.
