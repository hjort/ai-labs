## Imóveis residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv.bz2', sep=';')
dados.head()
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
dados.head(10)
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao.head(10)
```

```python
dados_residencial = dados[selecao]
dados_residencial.head(10)
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```
