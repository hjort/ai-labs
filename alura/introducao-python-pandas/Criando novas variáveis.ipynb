## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial2.csv', sep=';')
dados.head()
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head()
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head()
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head()
```

```python
dados['Tipo'].value_counts()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apto')
```

```python
dados.head(10)
```
